```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

class RCH_segment:
    pt_bottom_inside = (0,0)
    pt_bottom_outside = (0,0)
    pt_top_inside = (0,0)
    pt_top_outside = (0,0)    
    segment_type = ""
    fill = False
    color = (200,200,200)
    def __init__(self,pt_bottom_inside,pt_bottom_outside,pt_top_inside,pt_top_outside,segment_type, fill=False, color=(200,200,200)):
        self.pt_bottom_inside = pt_bottom_inside
        self.pt_bottom_outside = pt_bottom_outside
        self.pt_top_inside = pt_top_inside
        self.pt_top_outside = pt_top_outside
        self.segment_type = segment_type
        self.fill = fill
        self.color = color
    def get_points(self):
        return([self.pt_bottom_inside,self.pt_bottom_outside,self.pt_top_outside,self.pt_top_inside])
    def get_type(self):
        return(self.segment_type)


class RCH_diagram:
    # using inches since that is what the source uses, Can convert to metric after  
    inch_conp = 2
    inch_con_m = 39.37
    red = (255, 0, 0)
    gray = (200,200,200)
    black =  (0, 0, 0)
    white =  (255, 255, 255)
    lbrown =  (59, 92, 131)
    dbrown =  (40, 50, 63)
    green =  (40, 200, 40)
    height = int(17.5*12)
    width = 40*12
    origin_pixel_x = width * inch_conp /2
    origin_pixel_y = (height * inch_conp - 2)-100
    image_background = white
    sides = []
    sides.append("left")
    sides.append("right")
    segments = []

    rch_image = np.zeros((height * inch_conp, width * inch_conp, 3), np.uint8)
    color = tuple(reversed(white))
    # Fill image with color
    rch_image[:] = color
    chimney = [(-4,0),(4,132)]
    segments.append(RCH_segment((-5,0),((-15*12),0),(-5,48),((-13*12),48),"foot_left",False,black))
    segments.append(RCH_segment((5,0),((15*12),0),(5,48),((13*12),48),"foot_right",False,black))


    segments.append(RCH_segment((-5,49),((-13*12),49),(-5,96),((-9*12),96),"waist_left",False,black))
    segments.append(RCH_segment((5,49),((13*12),49),(5,96),((9*12),96),"waist_right",False,black))

    segments.append(RCH_segment((-5,97),((-9*12),97),(-5,120),((-6*12),120),"shoulder_left",False,black))
    segments.append(RCH_segment((5,97),((9*12),97),(5,120),((6*12),120),"shoulder_right",False,black))
    #segments.append(RCH_segment((-5,120),((-6*12),120),(-5,132),((-4.5*12),132),"head_left",True,lbrown))
    #segments.append(RCH_segment((5,120),((6*12),120),(5,132),((4.5*12),132),"head_right",True,lbrown))
    

    segments.append(RCH_segment((-15.25*12,0),((-17.25*12),0),(-13.25*12,48),((-13.75*12),48),"foot_cover_left",True,black))
    segments.append(RCH_segment((15.25*12,0),((17.25*12),0),(13.25*12,48),((13.75*12),48),"foot_cover_right",True,black))
    segments.append(RCH_segment((-13.25*12,48),((-13.75*12),48),(-9.25*12,96),((-9.75*12),96),"waist_cover_left",True,black))
    segments.append(RCH_segment((13.25*12,48),((13.75*12),48),(9.25*12,96),((9.75*12),96),"waist_cover_right",True,black))
    segments.append(RCH_segment((-9.25*12,96),((-9.75*12),96),(-6.25*12,120),((-6.75*12),120),"shoulder_cover_left",True,black))
    segments.append(RCH_segment((9.25*12,96),((9.75*12),96),(6.25*12,120),((6.75*12),120),"shoulder_cover_right",True,black))
    #segments.append(RCH_segment(((-6.25*12),120),((-6.75*12),120),((-3.25*12),138),((-3.75*12),138),"head_cover_left",True,black))
    #segments.append(RCH_segment(((6.25*12),120),((6.75*12),120),((3.25*12),138),((3.75*12),138),"head_cover_right",True,black))
    segments.append(RCH_segment((-5.25,127),((-5.65*12),127),(-5.25,138),((-3.75*12),138),"top_cover_left",True,black))
    segments.append(RCH_segment((5.25,127),((5.65*12),127),(5.25,138),((3.75*12),138),"top_cover_right",True,black))

    segments.append(RCH_segment((-15*12,0),((-15.25*12),0),(-13*12,48),((-13.25*12),48),"foot_leaf_cover_left",True,green))
    segments.append(RCH_segment((15*12,0),((15.25*12),0),(13*12,48),((13.25*12),48),"foot_leaf_cover_right",True,green))
    segments.append(RCH_segment((-13*12,48),((-13.25*12),48),(-9*12,96),((-9.25*12),96),"waist_leaf_cover_left",True,green))
    segments.append(RCH_segment((13*12,48),((13.25*12),48),(9*12,96),((9.25*12),96),"waist_leaf_cover_right",True,green))
    segments.append(RCH_segment((-9*12,96),((-9.25*12),96),(-6*12,120),((-6.25*12),120),"shoulder_leaf_cover_left",True,green))
    segments.append(RCH_segment((9*12,96),((9.25*12),96),(6*12,120),((6.25*12),120),"shoulder_leaf_cover_right",True,green))
    #segments.append(RCH_segment(((-6*12),120),((-6.25*12),120),(((-4.5*12)),132),(((-4.5*12)-3),132),"head_leaf_cover_left",True,green))
    #segments.append(RCH_segment(((6*12),120),((6.25*12),120),((4.5*12),132),((4.5*12+3),132),"head_leaf_cover_right",True,green))
    #segments.append(RCH_segment((-5,132),((-4.50*12),132),(-5,135),((-4.50*12+4.5),135),"top_leaf_cover_left",True,green))
    #segments.append(RCH_segment((5,132),((4.5*12),132),(5,135),((4.5*12+4.5),135),"top_leaf_cover_right",True,green))




    segments.append(RCH_segment((-17.25*12,-18),((17.25*12),-18),(-17.25*12,0),((17.25*12),0),"base",True,black))

    #segments.append(RCH_segment((-16*12,48),((-16*12),96),(-16*12-7,48),((-16*12-7),96),"billet",True,lbrown))
    segments.append(RCH_segment((-16*12-9,48),((-16*12-9),96),(-16*12-7-9,48),((-16*12-7-9),96),"billet",True,dbrown))
    
    segments.append(RCH_segment((-15*12-15,48),((-15*12-15),48+inch_con_m),(-15*12-1-15,48),((-15*12-1-15),48+inch_con_m),"m",True,black))
    segments.append(RCH_segment((16*12,48),((16*12),96),(16*12+2,48),((16*12+2),96),"lap-wood",True,lbrown))
    #segments.append(RCH_segment((16*12-4,48),((16*12-4),96),(16*12+2-4,48),((16*12+2-4),96),"lap-wood",True,dbrown))
    segments.append(RCH_segment((16*12+6,48),((16*12+6),48+inch_con_m),(16*12+1+6,48),((16*12+1+6),48+inch_con_m),"m",True,black))    



    def __init__(self):
        pass
        self.add_layer(4,120,(6*12),12,1.5,"head",True,self.lbrown,self.green)
        self.add_layer(4,132,(4.5*12),3,2,"head_cover",True,self.green,self.green)
        self.add_layer(4,135,(4.5*12)-6,12,2,"head_cap",True,self.black,self.black)
        self.segments.append(RCH_segment((-2*12,125),((-2*12),132),(2*12,125),((2*12),132),"Bridgen",True,self.dbrown))
    def convert_point(self, point_inches):
        p_x = int((point_inches[0] * self.inch_conp) + self.origin_pixel_x)
        p_y = int(self.origin_pixel_y - (point_inches[1] * self.inch_conp))
        return((p_x,p_y))
    def add_layer(self,x_origin, y_origin, width, height, slant, layer_name, layer_fill, layer_color, layer_color_2):
        for cn in range(0,2):
            if(cn==0):
                dir = -1
            else:
                dir = 1  
            #print(dir)  
            self.segments.append(RCH_segment((x_origin*dir,y_origin),((width*dir),y_origin),(x_origin*dir,y_origin+height),((width-(height*slant))*dir,y_origin+height),layer_name+"_"+self.sides[cn],layer_fill,layer_color))
            self.segments.append(RCH_segment((width*dir,y_origin),((width+3)*dir,y_origin),((width-(height*slant))*dir,y_origin+height),(((width-(height*slant))+3)*dir,y_origin+height),layer_name+"_"+self.sides[cn]+"_leaf",True,layer_color_2))
            self.segments.append(RCH_segment(((width+3)*dir,y_origin),((width+9)*dir,y_origin),(((width-(height*slant))+3)*dir,y_origin+height),(((width-(height*slant))+9)*dir,y_origin+height),layer_name+"_"+self.sides[cn]+"_dust",True,self.black))
            #segments.append(RCH_segment((5,120),((6*12),120),(5,132),((4.5*12),132),"head_right",True,lbrown))


    def convert_rect(self, points_inches):
        #print(points_inches[0])
        p1 = points_inches[0]
        p1px = self.convert_point(p1)
        p2 = points_inches[1]
        p2px = self.convert_point(p2)
        #print(p1px,p2px)
        return([p1px,p2px])

    def convert_points(self, points_inches):
        #print(points_inches[0])
        pts_px = []
        for p_in in points_inches:
            p_px = self.convert_point(p_in)
            p_px = [p_px[0],p_px[1]]
            pts_px.append(p_px)

        #print(pts_px)
        pts_px = np.array(pts_px)
        return(pts_px)


    def draw_rch(self):
        chimney_pixels = self.convert_rect(self.chimney)
        #print(chimney_pixels[0])
        #cv2.rectangle(self.rch_image, chimney_pixels[0], chimney_pixels[1], self.black, 1)
        #cv2.rectangle(self.rch_image, (300, 4), (300, 400), self.black, 2)
        #Chimney
        col = self.lbrown  
        cv2.line(self.rch_image,(self.convert_point((5,0))),(self.convert_point((5,48))),col,3)
        cv2.line(self.rch_image,(self.convert_point((-5,0))),(self.convert_point((-5,48))),col,3)
        cv2.line(self.rch_image,(self.convert_point((5,48))),(self.convert_point((5,96))),col,3)
        cv2.line(self.rch_image,(self.convert_point((-5,48))),(self.convert_point((-5,96))),col,3)

        for cn in range (0,25):
            #if(cn%2==0):
            col = self.dbrown  
            #else:
                
            cv2.line(self.rch_image,(self.convert_point((5+3+(cn*7),2))),(self.convert_point((5+3+(cn*6),47))),col,10)
            cv2.line(self.rch_image,(self.convert_point((-5-3+(cn*-7),2))),(self.convert_point((-5-3+(cn*-6),47))),col,10)
        for cn in range (0,25):            
            col = self.lbrown              

            cv2.line(self.rch_image,(self.convert_point((5+3+4+(cn*7),1))),(self.convert_point((5+3+2+(cn*6),47))),col,3)
            cv2.line(self.rch_image,(self.convert_point((-5-3-4+(cn*-7),1))),(self.convert_point((-5-3-2+(cn*-6),47))),col,3)            
        for cn in range (0,18):
            col = self.dbrown  
            cv2.line(self.rch_image,(self.convert_point((5+3+(cn*8.3),50))),(self.convert_point((5+3+(cn*6),95))),col,10)
            cv2.line(self.rch_image,(self.convert_point((-5-3+(cn*-8.3),50))),(self.convert_point((-5-3+(cn*-6),95))),col,10) 
        for cn in range (0,18):
            col = self.lbrown
            cv2.line(self.rch_image,(self.convert_point((5+3+4.5+(cn*8.3),49))),(self.convert_point((5+3+2+(cn*6),95))),col,3)
            cv2.line(self.rch_image,(self.convert_point((-5-3-4.5+(cn*-8.3),49))),(self.convert_point((-5-3-2+(cn*-6),95))),col,3) 


        for cn in range (0,24):
            #if(cn%2==0):
            col = self.dbrown  
            #else:
            #col = self.dbrown
            cv2.line(self.rch_image,(self.convert_point((8+(cn*4.2),97))),(self.convert_point((8+(cn*3),119))),col,10)
            cv2.line(self.rch_image,(self.convert_point((-8+(cn*-4.2),97))),(self.convert_point((-8+(cn*-3),119))),col,10) 

        for cn in range (0,24):
            #if(cn%2==0):
            col = self.lbrown  
            #else:
            #col = self.dbrown
            cv2.line(self.rch_image,(self.convert_point((4+1.5+(cn*4.2),97))),(self.convert_point((4+1.5+(cn*3),119))),col,3)
            cv2.line(self.rch_image,(self.convert_point((-4-1.5+(cn*-4.2),97))),(self.convert_point((-4-1.5+(cn*-3),119))),col,3) 



        #foot_l_px = self.convert_points(self.foot_left)
        #cv2.polylines(self.rch_image,[foot_l_px],True,self.black)
        #foot_r_px = self.convert_points(self.foot_right)
        #cv2.polylines(self.rch_image,[foot_r_px],True,self.black)
        for s in self.segments:
            #print(s.get_type(), s.fill)
            #print(s.get_points())            
            if(s.fill==True):
                cv2.fillPoly(self.rch_image,[self.convert_points(s.get_points())],s.color)
                
            else:
                pass  
                #cv2.polylines(self.rch_image,[self.convert_points(s.get_points())],True,s.color)
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = .6
        font_color = self.black 
        font_thickness = 1
        cv2.rectangle(self.rch_image, (self.convert_point((-72,-16))), (self.convert_point((200,-6))), self.white, -1)
        cv2.putText(self.rch_image, "Base (Smooth, cleared of vegetation.) ~10m diameter." , (self.convert_point((-72,-14))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.rectangle(self.rch_image, (self.convert_point((12,2))), (self.convert_point((35,12))), self.white, -1)
        cv2.putText(self.rch_image, "Foot" , (self.convert_point((12,4))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.rectangle(self.rch_image, (self.convert_point((12,50))), (self.convert_point((37,60))), self.white, -1)
        cv2.putText(self.rch_image, "Waist" , (self.convert_point((12,52))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.rectangle(self.rch_image, (self.convert_point((12,98))), (self.convert_point((60,108))), self.white, -1)
        cv2.putText(self.rch_image, "Shoulders" , (self.convert_point((12,100))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.rectangle(self.rch_image, (self.convert_point((30,122))), (self.convert_point((56,132))), self.white, -1)
        cv2.putText(self.rch_image, "Head" , (self.convert_point((30,124))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)

        cv2.rectangle(self.rch_image, (self.convert_point((-90,62))), (self.convert_point((-20,72))), self.white, -1)
        cv2.putText(self.rch_image, "Kindling, chips" , (self.convert_point((-90,64))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.arrowedLine(self.rch_image,(self.convert_point((-20,67))),(self.convert_point((-4,67))),self.white,2,tipLength = 0.4)

        cv2.putText(self.rch_image, "Billet, meter" , (self.convert_point((-17.5*12,98))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.putText(self.rch_image, "Lap-wood, meter" , (self.convert_point((10*12,98))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)

        
        cv2.putText(self.rch_image, "Chimney" , (self.convert_point((-24,150))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.putText(self.rch_image, "Bridgen" , (self.convert_point((-24,126))), font, .5, self.white, font_thickness, cv2.LINE_AA)
        cv2.putText(self.rch_image, "Dust and soil cover" , (self.convert_point((-13*12,150))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.putText(self.rch_image, "Leaf cover layer" , (self.convert_point((3*12,150))), font, font_scale, font_color, font_thickness, cv2.LINE_AA)
        cv2.arrowedLine(self.rch_image,(self.convert_point((4*12-6,148))),(self.convert_point((4*12-6,136))),self.green,2,tipLength = 0.4)
        cv2.arrowedLine(self.rch_image,(self.convert_point((-8*12,148))),(self.convert_point((-8*12,112))),self.black,2,tipLength = 0.2)

        
        for cn in range(self.origin_pixel_y-226,self.origin_pixel_y,6):
        #for cn in range(200,300):
            # print(int(self.origin_pixel_x-120)))
            self.rch_image[cn:cn+1,int(self.origin_pixel_x-7):int(self.origin_pixel_x+3)] = self.lbrown
            self.rch_image[cn:cn+1,int(self.origin_pixel_x+5):int(self.origin_pixel_x+7)] = self.lbrown
            self.rch_image[cn+3:cn+4,int(self.origin_pixel_x-7):int(self.origin_pixel_x-5)] = self.lbrown
            self.rch_image[cn+3:cn+4,int(self.origin_pixel_x-3):int(self.origin_pixel_x+7)] = self.lbrown

            #self.rch_image[cn:cn+1,int(self.origin_pixel_x-8):int(self.origin_pixel_x+8)] = [25, 255, 255]
            #self.rch_image[cn:cn+5,int(300-8):int(300+8)] = [25, 255, 255]


        return(self.rch_image)
rch = RCH_diagram()
image = rch.draw_rch()
cv2_imshow(image)
```
